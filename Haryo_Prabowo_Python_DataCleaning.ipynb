# 1. Location Format Check and Rename

```python
def loc_check(city):
  x = city.split(', ')
  if len(x) == 1 or len(x[1]) != 2 or (x[1].upper() != x[1]): #harus ada 2 huruf kapital setelah koma
    return 'others'
  else:
    return city

df['Location'] = df['Location'].apply(loc_check)
df
```

Semua lokasi yang tidak mengikuti format "City, State Code" sudah diubah menjadi "others".
